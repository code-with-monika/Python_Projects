```python
from tkinter import *
import cv2
import threading
from datetime import datetime
import requests

cap = cv2.VideoCapture(0)

def get_latitude_from_ip():
    try:
        response = requests.get("https://ipinfo.io/json")
        data = response.json()
        if "loc" in data:
            latitude = data["loc"].split(",")[0]  
            return float(latitude)
        else:
            return None
    except Exception as e:
        print("Error retrieving latitude:", e)
        return None

def capture_image():
    img_counter = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
            
        cv2.imshow('Frame', frame)
        key = cv2.waitKey(1)

        if key == ord('q'):
            break
        elif key % 256 == 32:  
           
            frame_with_timestamp = frame.copy()
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            cv2.putText(frame_with_timestamp, timestamp, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 
                        0.7, (0, 255, 0), 2, cv2.LINE_AA)
        
          
            latitude = get_latitude_from_ip()
            if latitude is not None:
                latitude_text = f"Latitude: {latitude}"
                cv2.putText(frame_with_timestamp, latitude_text, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 
                            0.7, (0, 255, 0), 2, cv2.LINE_AA)
               # print(f"Latitude: {latitude}")
            else:
                print("Can not retrive latitude.")
                
           
            img_name = "Monika_{}.jpg".format(img_counter)
            cv2.imwrite(img_name, frame_with_timestamp)
            print("Image Captured")
            
            img_counter += 1
    
    cap.release()
    cv2.destroyAllWindows()

def start_capture():
    threading.Thread(target=capture_image).start()
    
window = Tk()
window.geometry("300x300")
window.title("GPS Camera")
label = Label(window, text="Geo Tagged Image Capture", font=("Bold", 18), fg="Black")
label.pack() 
button = Button(window, text="Capture Image", command=start_capture)
button.pack()

window.mainloop()
```
